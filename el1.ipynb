```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks = pd.read_csv("tracks.csv")
genre = pd.read_csv("SpotifyFeatures.csv")
```

```python
tracks.head()
```

```python
genre.head()
```

```python
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
# Cleaning the dataset
tracks.dropna(subset=["name"], inplace=True)
```

```python
tracks.info()
```

```python
tracks.describe().transpose()
```

```python
# Top 15 most popular songs on Spotify
popular = tracks.query("popularity > 90", inplace=False).sort_values('popularity', ascending=False)
popular[:15]
```

```python
# 10 least popular songs
least = tracks.sort_values(by="popularity", ascending=True).head(10)
least
```

```python
# Earliest songs available on Spotify
tracks.set_index("release_date", inplace=True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head(10)
```

```python
tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks.duration.head(10)
```

```python
# Karl Pearsons Correlation Coefficient between all parameters
actual = tracks.drop(labels=["explicit", "key", "mode"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
map = sns.heatmap(actual, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Pearsons Correlation Heatmap between All Variables')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
# Spearman Correlation Coefficient between all parameters
actual1 = tracks.drop(labels=["explicit", "key", "mode"], axis=1).corr(method="spearman")
plt.figure(figsize=(14,6))
map = sns.heatmap(actual1, annot = True, fmt = '.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
map.set_title('Spearmans Correlation Heatmap between All Variables')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
#Now we will consider 40% of the dataset to plot regression Lines
# This is because the original dataset is too large, and graphs will become overpopulated
```

```python
sam = tracks.sample(int(0.004 * len(tracks)))
len(sam)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='popularity', x='acousticness', color='b').set(title="Popularity vs Acousticness Correlation")
```

```python
# Here, we create a new column called year to plot the number of songs released yearly
tracks['dates']=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
# To observe the change in songs length over the years
total_dr = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title='Years vs Duration')
plt.xticks(rotation=90)
```

```python
# Duration of songs in different genres analysis
plt.title('Duration of songs in different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
# Top 5 genres by popularity
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
popular = genre.sort_values('popularity', ascending=False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = popular).set(title='Top 5 Genres by Popularity')
```
